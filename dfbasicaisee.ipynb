````python
import requests
from bs4 import BeautifulSoup
import json

# Define URL and headers
url = "https://www.overleaf.com/learn/latex/Learn_LaTeX_in_30_minutes"
headers = {"User-Agent": "Mozilla/5.0"}

# Request page content
response = requests.get(url, headers=headers)
if response.status_code != 200:
    print("Failed to retrieve the page.")
    exit()

# Parse HTML
soup = BeautifulSoup(response.text, "html.parser")

# Extract structured content
content_div = soup.find("div", {"class": "mw-parser-output"})
if not content_div:
    print("Failed to find content section.")
    exit()

# Store structured corpus
corpus = []
current_section = None

for element in content_div.find_all(["h2", "h3", "p", "ul", "ol", "pre"]):
    if element.name in ["h2", "h3"]:
        # Create new section
        current_section = {"title": element.get_text(strip=True), "content": []}
        corpus.append(current_section)
    elif element.name == "p":
        # Append paragraph text
        if current_section:
            current_section["content"].append(element.get_text(strip=True))
    elif element.name in ["ul", "ol"]:
        # Append list items
        if current_section:
            items = [li.get_text(strip=True) for li in element.find_all("li")]
            current_section["content"].extend(items)
    elif element.name == "pre":
        # Append LaTeX code blocks
        if current_section:
            current_section["content"].append(f"```latex\n{element.get_text(strip=True)}\n```")

# Save corpus as JSON
with open("latex_corpus.json", "w", encoding="utf-8") as f:
    json.dump(corpus, f, ensure_ascii=False, indent=4)

print(f"✅ Extracted {len(corpus)} sections and saved as latex_corpus.json")
````
